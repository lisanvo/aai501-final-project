```python
for index, row in metadata.iterrows():
    if index > filtered_test.shape[0] - 1:
        break
        
    file_loc = row['File Location']
    subject_id = filtered_test.loc[index, 'Subject ID']
    study_uid = filtered_test.loc[index, 'Study UID']
    series_uid = filtered_test.loc[index, 'Series UID']
    
    if subject_id == row['Subject ID'] and study_uid == row['Study UID'] and series_uid == row['Series UID']:

        jpg_folder_path = '../image_sets/train/' + filtered_test.loc[index, 'pathology']
        sliced_file_loc = file_loc[2:]
        source_path = prefix + sliced_file_loc
        
        images_path = os.listdir(source_path)
        for n, image in enumerate(images_path):
            ds = dicom.dcmread(os.path.join(source_path, image))
            pixel_array_numpy = ds.pixel_array
            image = image.replace('.dcm', '.jpg')
            cv2.imwrite(os.path.join(jpg_folder_path, image), pixel_array_numpy)
            os.rename(jpg_folder_path + "/" + image, jpg_folder_path + "/" + str(index) + ".jpg")
```

# Pre-Processing

```python
train_path = '../image_sets/train'
validation_path = '../image_sets/validation'
test_path = '../image_sets/test'
```

```python
train_batches = ImageDataGenerator().flow_from_directory(train_path, target_size=(224,224), classes=['BENIGN', 'MALIGNANT', 'BENIGN_WITHOUT_CALLBACK'], batch_size=10)
validation_batches = ImageDataGenerator().flow_from_directory(validation_path, target_size=(224,224), classes=['BENIGN', 'MALIGNANT', 'BENIGN_WITHOUT_CALLBACK'], batch_size=4)
test_batches = ImageDataGenerator().flow_from_directory(test_path, target_size=(224,224), classes=['BENIGN', 'MALIGNANT', 'BENIGN_WITHOUT_CALLBACK'], batch_size=10)
```

```python
imgs, labels = next(train_batches)
```

# Create and train CNN model

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    Flatten(),
    Dense(3, activation='softmax')
])
```

```python
model.compile(Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_batches, steps_per_epoch=4, validation_data=validation_batches, validation_steps=4, epochs=5, verbose=2)
```

# Predict

```python
predictions = model.predict_generator(test_batches, steps=1, verbose=0)
```
